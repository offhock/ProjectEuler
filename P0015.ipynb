````python
# 조합의 수를 구하는 것과 같다.
# 예를 들어 , 3x3의 경우, 경로가 6단계로 구성되므로,
# 6개 중에 3개를 뽑는 경우의 수와 동일하다.
# nCr = n! / (r! * (n-r)!)

def factorial(x):
    mul=1
    for i in range(1,x+1):
        mul=mul*i
    return mul
    
def C(n,r):
    tmp = factorial(n) / (factorial (r) * factorial (n-r))
    return tmp

print(C(1+1,1))  # 1x1
print(C(2+2,2))  # 2x2
print(C(3+3,3))  # 3x3
print(C(4+4,4))  # 4x4
print(C(10+10,10)) # 10x10
print(C(20+20,20)) # 20x20
````

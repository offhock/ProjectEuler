# Number letter counts


Problem 17
If the numbers 1 to 5 are written out in words: one, two, three, four, five, then there are $3 + 3 + 5 + 4 + 4 = 19$ letters used in total.

If all the numbers from 1 to 1000 (one thousand) inclusive were written out in words, how many letters would be used?


> NOTE: Do not count spaces or hyphens. For example, 342 (three hundred and forty-two) contains 23 letters and 115 (one hundred and fifteen) contains 20 letters. The use of "and" when writing out numbers is in compliance with British usage.


>Answer:  21124  
> Completed on Tue, 9 Aug 2022, 17:18

```python
str_1 = ['','one','two','three','four','five','six','seven','eight','nine']
str_10 =['ten','eleven','twelve','thirteen','fourteen','fifteen','sixteen','seventeen','eighteen','nineteen']  
str_20 = ['','ten','twenty','thirty','forty','fifty','sixty','seventy','eighty','ninety']
str_100 = ['','hundred','hundred','hundred','hundred','hundred','hundred','hundred','hundred','hundred','hundred']
str_1000 = ['','thousand','thousand','thousand','thousand','thousand','thousand','thousand','thousand','thousand']

count =0
space = ' '
for i in range(1,1001):    
    
    p4 = int(i/1000)
    p3 = int((i%1000)/100)
    p2 = int(((i%1000)%100)/10)
    p1 = int(((i%1000)%100)%10)
          
    if p4==0 and  p3 == 0 or p2 ==0 and p1 ==0:
        s_and = ' '        
    else:
        s_and = ' and '
    
    s = str_1[p4] + space + str_1000[p4] 
    s = s + space+ str_1[p3] +space+ str_100[p3]  + s_and
    if (i%1000)%100>=11 and (i%1000)%100<20:
        s = s + space +  str_10[p1] 
    else:       
        s = s + space +  str_20[p2] 
        s = s + space+ str_1[p1]
    l =  len(s) - s.count(' ')
    count = count + l
    print(i, p4,p3,p2,p1, s, l)    

print (count)     
```
